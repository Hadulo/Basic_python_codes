# Numpy

```python
import numpy as np

 

# Define a structured data type

dtype = np.dtype([('name', 'U10'), ('age', 'i4'), ('weight', 'f4')])

 

# Create a structured array

data = np.array([('Alice', 25, 55.5), ('Bob', 30, 85.0), ('Cathy', 22, 45.8)], dtype=dtype)

 

# Access the entire array

print("Structured Array:\n", data)

 

# Access specific fields

print("\nNames:", data['name'])

print("Ages:", data['age'])

print("Weights:", data['weight'])
```

```python
import numpy as np

 

# Define a structured data type

dtype = np.dtype([('name', 'U10'), ('age', 'i4'), ('weight', 'f4')])

 

# Create a structured array first

data = np.array([('Alice', 25, 55.5), ('Bob', 30, 85.0), ('Cathy', 22, 45.8)], dtype=dtype)

 

# Convert structured array to a record array

rec_data = data.view(np.recarray)

 

# Access the entire array

print("Record Array:\n", rec_data)

 

# Access specific fields using both dictionary-style and attribute-style

print("\nNames (dictionary-style):", rec_data['name'])

print("Names (attribute-style):", rec_data.name)

print("Ages (dictionary-style):", rec_data['age'])

print("Ages (attribute-style):", rec_data.age)

print("Weights (dictionary-style):", rec_data['weight'])

print("Weights (attribute-style):", rec_data.weight)
```
